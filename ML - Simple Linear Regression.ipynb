We will be doing Simple Linear Regression on salary data. Very simple example. 

We will train a model that understands the correlation between years of 
experience and salary. we will obtain predictions for salary (y)

Note that for SLR, we dont have to worry about feature scaling as the library will take care of this for us.

It is a good habit to have your dataset structured appropriatly before importing into pandas. That is to say, put your dependent variable as the last column.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values #important that you do it with the -1
#dataset.iloc[:, 0].values -- creates an issue later on with matrix manipulation
y = dataset.iloc[:, 1].values

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, 
                                                    random_state = 0)
# we use random state here so you obtain identical results.

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

###### we will now predict the test set results.
y_pred = regressor.predict(X_test) #this variable stores the predicted values.

## we then compare y_pred (what our model predicted) 
# with y_test(true values we know) via visualization

####visualization time - training set
plt.scatter(X_train, y_train, color='red') #shows real values of experience and salary
plt.plot(X_train, regressor.predict(X_train), color='blue') #compares real values of experience and predicted salary
plt.title('Salary vs. Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

#red points are the real values
#blue line is the prediction
#RECALL THAT THIS IS THE SLR ON THE TRAINING MODEL!
#lets see how this blue line fairs on the test set!

#visualization time - test set
plt.scatter(X_test, y_test, color='red') #shows real values of experience and salary in the test set.
plt.plot(X_train, regressor.predict(X_train), color='blue') #compares real values of experience and predicted salary
plt.title('Salary vs. Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
#the second line plt.plot doesnt need to be changed at all
#we trained our model on the training set.
#thus we are assessing how our model fares on the test data.
```

From the above picture, we can see that our model is quite good at predicting the test case!

That's all there is to simple linear regression using the sklearn package.
